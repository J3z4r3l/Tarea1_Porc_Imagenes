```python
#Agregamos las librerias
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os
import cv2
```

```python
def grey_image(source_color, source_b_w):
    # Obtenemos la lista de archivos de la carpeta
    archivos_carpeta = os.listdir(source_color)
    # Iteramos sobre cada archivo y agregamos solo los archivos de imagen a la lista
    Image_name = [archivo for archivo in archivos_carpeta if archivo.endswith(('.png', '.jpg', '.jpeg'))]
    # Iteramos sobre cada imagen
    for i, nombre_imagen in enumerate(Image_name):
        # Creamos la ruta completa del archivo de imagen
        ruta_imagen = os.path.join(source_color, nombre_imagen)
        Color_image = cv2.imread(ruta_imagen)
        # Convertimos la imagen a escala de grises
        imagen_gris = cv2.cvtColor(Color_image, cv2.COLOR_BGR2GRAY)
        # Generamos el nombre del archivo de salida y lo guardamos
        file_name = f"{os.path.splitext(nombre_imagen)[0]}_{i+1}_bn.png"
        ruta_archivo_gris = os.path.join(source_b_w, file_name)
        cv2.imwrite(ruta_archivo_gris, imagen_gris)
```

```python
def binarize_image(source_b_w, source_binary, threshold=100):
    # Obtener la lista de archivos de la carpeta
    archivos_carpeta_bn = os.listdir(source_b_w)
    nombres_imagenes_bn = []
    numero_cambios_bn = 1

    # Binarización de las imágenes
    for archivo_bn in archivos_carpeta_bn:
        # Comprobamos si el archivo tiene una extensión de imagen válida
        # Agregamos el nombre del archivo a la lista de nombres de imágenes
        nombres_imagenes_bn.append(archivo_bn)
            
    for imagenes_bn in nombres_imagenes_bn:
       # Creamos la ruta completa del archivo de imagen
        source_img_by = os.path.join(source_b_w, imagenes_bn)
        img_bn = cv2.imread(source_img_by)
        # Aplicar umbralización
        umbralizado_v0=img_bn.copy()
        umbralizado_v0[umbralizado_v0 < threshold] = 0
        umbralizado_v0[umbralizado_v0 >= threshold] = 255
        nombre_archivo_bn = os.path.splitext(imagenes_bn)[0]
        nuevo_nombre_archivo_bn = f"{nombre_archivo_bn}_{numero_cambios_bn}_binario.png"
        ruta_archivo_modificado_blanco = os.path.join(source_binary, nuevo_nombre_archivo_bn)
        # Guardamos la imagen binarizada
        cv2.imwrite(ruta_archivo_modificado_blanco, umbralizado_v0)
        numero_cambios_bn += 1
```

```python
def binarize_image_c(ruta_color, ruta_binario_color, umbral=127):
    # Obtener la lista de archivos de la carpeta
    archivos_carpeta_binario_color = os.listdir(ruta_color)
    # Binarización de las imágenes
    for i, archivo in enumerate(archivos_carpeta_binario_color):
        # Creamos la ruta completa del archivo de imagen
        ruta_imagen = os.path.join(ruta_color, archivo)
        # Cargamos la imagen y aplicamos umbralización para la imagen a color
        imagen = cv2.imread(ruta_imagen)
        binarizado = cv2.threshold(imagen[:, :, 1], umbral, 255, cv2.THRESH_BINARY)[1]
        # Generamos el nombre del archivo de salida y lo guardamos
        nuevo_nombre = f"{os.path.splitext(archivo)[0]}_{i+1}_bincol.png"
        ruta_archivo_modificado = os.path.join(ruta_binario_color, nuevo_nombre)
        cv2.imwrite(ruta_archivo_modificado, binarizado)
```

```python
def gaussian_noise(source_folder, destination_folder, kernel_size=5):
    # Obtener la lista de archivos de la carpeta
    files_list = os.listdir(source_folder)
    image_names = []
    # Iteramos sobre cada archivo y agregamos solo los archivos de imagen a la lista
    for file_name in files_list:
        # Agregamos el nombre del archivo a la lista de nombres de imagenes
        image_names.append(file_name)

    # Iteramos sobre cada imagen
    for image_name in image_names:
        # Creamos la ruta completa del archivo de imagen
        image_path = os.path.join(source_folder, image_name)
        image = cv2.imread(image_path)

        # Aplicamos el filtro de ruido gaussiano
        noisy_image = cv2.GaussianBlur(image, (kernel_size, kernel_size), 0)

        # Creamos el nombre del archivo de salida y la ruta completa
        new_name = f"{os.path.splitext(image_name)[0]}_gaussian.png"
        destination_path = os.path.join(destination_folder, new_name)

        # Guardamos la imagen con ruido gaussiano en la carpeta de destino
        cv2.imwrite(destination_path, noisy_image)
```

```python
#Suavizado
def suavizado(source_folder, destination_folder):
    # Obtener la lista de archivos de la carpeta
    files_list = os.listdir(source_folder)
    image_names = []
    # Iteramos sobre cada archivo y agregamos solo los archivos de imagen a la lista
    for file_name in files_list:
        # Agregamos el nombre del archivo a la lista de nombres de imagenes
        image_names.append(file_name)

    # Iteramos sobre cada imagen
    for image_name in image_names:
        # Creamos la ruta completa del archivo de imagen
        image_path = os.path.join(source_folder, image_name)
        image = cv2.imread(image_path)

        # Aplicamos el filtro de ruido gaussiano
        img_filtrada = cv2.medianBlur(image, 3)

        # Creamos el nombre del archivo de salida y la ruta completa
        new_name = f"{os.path.splitext(image_name)[0]}_suav.png"
        destination_path = os.path.join(destination_folder, new_name)

        # Guardamos la imagen con ruido gaussiano en la carpeta de destino
        cv2.imwrite(destination_path, img_filtrada)
```

```python
def bordes(source_folder, destination_folder):
    # Obtener la lista de archivos de la carpeta
    files_list = os.listdir(source_folder)
    image_names = []
    # Iteramos sobre cada archivo y agregamos solo los archivos de imagen a la lista
    for file_name in files_list:
        # Agregamos el nombre del archivo a la lista de nombres de imagenes
        image_names.append(file_name)

    # Iteramos sobre cada imagen
    for image_name in image_names:
        # Creamos la ruta completa del archivo de imagen
        image_path = os.path.join(source_folder, image_name)
        image = cv2.imread(image_path)

        # Aplicamos el filtro de ruido gaussiano
        canny = cv2.Canny(image, 100, 200)

        # Creamos el nombre del archivo de salida y la ruta completa
        new_name = f"{os.path.splitext(image_name)[0]}_BOR.png"
        destination_path = os.path.join(destination_folder, new_name)

        # Guardamos la imagen con ruido gaussiano en la carpeta de destino
        cv2.imwrite(destination_path, canny)
```
